```python
# Create a new column math_level:

# High: ≥ 85

# Medium: 60–84

# Low: < 60

# Then group by math_level and calculate the mean of all three scores.
```

```python
def grade_cal(r):
    if r['math score']>85:
        return 'High'
    elif 60<r['math score']<84:
        return 'Medium'
    elif r['math score']<60:
        return 'Low'
    
df['math_level']=df.apply(grade_cal,axis = 1)
mean = df.groupby('math_level')[['math score', 'reading score', 'writing score']].mean()
mean
```

```python
df['math_level'] = pd.cut(
    df['math score'],
    bins=[-1, 59, 84, 100],
    labels=['Low', 'Medium', 'High']
)
```

```python
math_level_mean = df.groupby('math_level')[['math score', 'reading score', 'writing score']].mean()
math_level_mean
```

```python
# Group by test preparation course and:

# compute the mean of all 3 scores

# compute the percentage improvement in math scores for students who completed the course vs those who didn’t
```

```python
prep_means = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
prep_means
```

```python
completed_mean = prep_means.loc['completed', 'math score']
none_mean = prep_means.loc['none', 'math score']

percentage_improvement = ((completed_mean - none_mean) / none_mean) * 100
percentage_improvement
```

```python
# For each parental education level:
# Compute the mean total score
# (total = math + reading + writing).

# Then return only the groups whose mean total score is ≥ 220.
```

```python
df['total'] = df['math score'] + df['reading score'] + df['writing score']

parent_total = df.groupby('parental level of education')['total'].mean()

parent_total_filtered = parent_total[parent_total >= 220]
parent_total_filtered
```
