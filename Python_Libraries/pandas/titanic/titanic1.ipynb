```python
def comp_type(row):
    if (row['psg_cnt'] == 1) and row['fam_count']==0:
        return 'Solo'
    elif row['psg_cnt'] -1 == row['fam_count']:
        return 'Family'
    elif row['psg_cnt']>1 and  row['fam_count']==0:
        return 'Friends'
    elif row['psg_cnt']>row['fam_count']:
        return 'Family and Friends'
    else:
        return 'Family'
df['companion_type'] =df.apply(comp_type,axis=1)
```

```python
def social_status(row):
    if row['courtesy_title'] in ['Miss','Mrs','Mr','Ms','Mlle','Mme','Master']:
        return 'Common'
    elif row['courtesy_title'] in ['Lady','Jonkheer','The Countess','Don','Dona','Sir']:
        return 'Royal'
    elif row['courtesy_title'] in ['Major','Col','Capt']:
        return 'Military'
    else:
        return 'Other Profession'

df['social_status']=df.apply(social_status,axis=1)    
```

```python
Fare = df['fare']


Q1 = np.percentile(df['fare'], 25)
Q3 = np.percentile(df['fare'], 75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = df['fare'][(df['fare']< lower_bound) | (df['fare'] > upper_bound)]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers in Fare Column:")
print(outliers)

```
