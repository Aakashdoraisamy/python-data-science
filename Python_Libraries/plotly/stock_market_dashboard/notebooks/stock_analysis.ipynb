```python
import yfinance as yf
import pandas as pd
import plotly.express as px

ticker = "AAPL"
start_date = "2023-01-01"
end_date = "2023-12-31"

# Try to fetch data from Yahoo Finance
try:
    df = yf.download(ticker, start=start_date, end=end_date, progress=False, timeout=10)
    df.reset_index(inplace=True)
    
    # Check if data was fetched
    if df.empty:
        print("No data fetched. Please check internet connection or try a smaller date range.")
    else:
        # Closing price chart
        fig = px.line(df, x='Date', y='Close', title=f"{ticker} Closing Price")
        fig.show()

        # Volume chart
        fig2 = px.bar(df, x='Date', y='Volume', title=f"{ticker} Trading Volume")
        fig2.show()

except Exception as e:
    print(f"Error fetching {ticker}: {e}")
```
