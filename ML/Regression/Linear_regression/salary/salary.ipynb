```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plott
import seaborn as sns
```

```python
# weights are importance of features
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain) # y = mx+c
```

```python
print("Slope:", model.coef_)
```

```python
print("Intercept:", model.intercept_)
```

```python
ypred = model.predict(xtrain)
```

```python
plott.scatter(df['YearsExperience'],df['Salary'])
```

```python
sns.scatterplot(x='YearsExperience', y='Salary',data=df)
```

```python
plott.scatter(xtrain['YearsExperience'],ytrain)
plott.scatter(xtrain['YearsExperience'],ypred,marker='*')
```

```python
ypred_t = model.predict(xtest)
plott.scatter(xtest['YearsExperience'],ytest)
plott.scatter(xtest['YearsExperience'],ypred_t,marker='*')
```

## interview question

### what is metrics ? 
### metrics which is module which contains different evaluation metrics for regression and classification tasks. It provides functions to measure the performance of machine learning models by comparing predicted values with actual values

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(ytrain, ypred)
```

```python
rmse = root_mean_squared_error(ytrain,ypred)
(rmse/ytrain.mean())*100
```

```python
mean_squared_error(ytest, ypred_t)
```

```python
print("R2 Score on test:", r2_score(ytest, ypred_t))
```

```python
print('R2 Score on train',r2_score(ytrain,ypred))
```

```python
model.score(xtrain,ytrain)
```
