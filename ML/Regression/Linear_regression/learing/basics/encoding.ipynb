### drawbacks of LabelEncoder
- It assigns arbitrary numbers to categories which may mislead some algorithms to assume an order or priority among categories.
- It is not suitable for nominal categorical variables with no inherent order.
- It can lead to the curse of dimensionality if there are many unique categories, resulting in a large number of features.
- if there is more unique categories it may not be efficient.

### One-Hot Encoding

```python
df = pd.DataFrame({'product':['watch','laptop','tab','laptop','tab','tab','watch']})
df
```

```python
ohe = OneHotEncoder(sparse_output=False) # instance of class
# Instead of storing all those zeros, we store only where 1 exists.
# spare matrix is memeory efficient way of storing data.
```

```python
data = ohe.fit_transform(df[['product']])
```

```python
col_name = ohe.get_feature_names_out(['product']) # getting column names after one hot encoding
col_name
```

```python
df1 = pd.DataFrame(data,columns=col_name)
df1
```
