# One Hot Encoding

```python
encoder = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
encoded_data = encoder.fit_transform(data)
encoded_data
```

```python
encoder.transform([['S1','Good1']])
```

```python
encoder.get_feature_names_out()
pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out())
```

# Ordinal Encoding

```python
data=pd.DataFrame({'grade':['S','A','S','A','F'],'exam_result':['Excellent','Good','Excellent','Good','Fail']})
data
```

```python
ord_encoder = OrdinalEncoder(categories=[['S','A','F'],['Excellent','Good','Fail']])
```

```python
ord_encoder.fit_transform(data) # accepts only 2D
```

```python
ord_encoder.fit_transform([['S','Good']])
```

```python
ord_encoder = OrdinalEncoder()
ord_encoder.fit_transform([['S1','Good']])
```

```python
ord_encoder = OrdinalEncoder(categories=[['S','A','F'],['Excellent','Good','Fail']],handle_unknown='use_encoded_value',unknown_value=-1)
# here unknown categories will be encoded as -1 and we pass categories in order to avoid error
```

```python
ord_encoder.fit_transform(data) # learns pattern from data
```

```python
ord_encoder.transform([['W','Average']])
```
