```python
import pandas as pd

train_data = pd.read_csv("Google_Stock_Train.csv")
test_data = pd.read_csv("Google_Stock_Test.csv")

print(train_data.head())
print(test_data.head())
```

```python
import datetime

# Convert date to ordinal (numeric)
train_data["DateOrd"] = pd.to_datetime(train_data["Date"]).map(datetime.datetime.toordinal)
test_data["DateOrd"] = pd.to_datetime(test_data["Date"]).map(datetime.datetime.toordinal)

# Assign X and y
X_train = train_data["DateOrd"].values.reshape(-1, 1)
y_train = train_data["Close"].values

X_test = test_data["DateOrd"].values.reshape(-1, 1)
y_test = test_data["Close"].values
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, color="blue")
plt.xlabel("Date (numeric)")
plt.ylabel("Close Price")
plt.title("Google Stock Price (Training Data)")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

r2_train = model.score(X_train, y_train)
print("R² (Training):", r2_train)
```
