```python
df = df.drop(columns=['homepage', 'tagline'])
```

```python
df['runtime'] = df['runtime'].fillna(df['runtime'].median())
```

```python
df['release_date'] = df['release_date'].fillna("Unknown")
```

```python
df['overview'] = df['overview'].fillna("NA")
```

```python
df.isnull().sum()
```

```python
df[['budget', 'revenue', 'runtime', 'vote_count', 'popularity']].describe()
```

```python
(df[['budget', 'revenue', 'runtime']] <= 0).sum()
```

```python
df = df[
    (df['budget'] > 0) &
    (df['revenue'] > 0) &
    (df['runtime'] > 0)
]
```

```python
df.shape
```

```python
df[['budget', 'revenue', 'runtime']].describe()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
df['release_date'].dtype
```

```python
df['release_date'].isnull().sum()
```

```python
df['release_year'] = df['release_date'].dt.year
```

```python
df[['release_date', 'release_year']].head()
```

```python
df = df.drop(columns=['release_date'])
```

```python
df.columns
```

```python
df = df.drop(columns=['overview', 'title', 'original_title',
                      'production_companies', 'production_countries', 'spoken_languages', 'keywords'])
```

```python
df.columns
```

```python
import ast  # to parse string representation of list/dict

def count_genres(x):
    try:
        genres_list = ast.literal_eval(x)
        return len(genres_list)
    except:
        return 0

df['num_genres'] = df['genres'].apply(count_genres)
```

```python
df[['genres', 'num_genres']].head()
```

```python
df = df.drop(columns=['genres'])
```

```python
print(df['status'].unique())
print(df['original_language'].unique())
```

```python
df['status'] = df['status'].map({'Released': 1, 'Post Production': 0})
```

```python
df['status'].unique()
```

```python
df['original_language'].value_counts()
```

```python
top_languages = ['en', 'fr', 'es', 'ja', 'zh', 'de']
df['original_language'] = df['original_language'].apply(lambda x: x if x in top_languages else 'Other')
```

```python
df = pd.get_dummies(df, columns=['original_language'], drop_first=True)
```

```python
df.head()
```

```python
df = df.drop(columns=['id'])
```

```python
df.columns
```

```python
X = df.drop(columns=['revenue'])  # features
y = df['revenue']                 # target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size = 0.8, random_state=42
)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X_train, y_train)
```

```python
lr_model.coef_
```

```python
y_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R² Score:", r2)
print("Mean Absolute Error:", mae)
```

```python
# Intercept
print("Intercept:", lr_model.intercept_)

# Coefficients (slopes) for each feature
for feature, coef in zip(X_train.columns, lr_model.coef_):
    print(f"{feature}: {coef}")
```
