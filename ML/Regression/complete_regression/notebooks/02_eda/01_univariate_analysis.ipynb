# Exploratory Data Analysis: Univariate Analysis

## Objective
The objective of this notebook is to analyze individual features to understand their distributions, spread, and potential anomalies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/processed/housing_cleaned.csv")
df.head()
```

```python
numerical_features = df.select_dtypes(include=[np.number]).columns
numerical_features
```

```python
for col in numerical_features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in numerical_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
sns.countplot(y=df["ocean_proximity"])
plt.title("Distribution of Ocean Proximity Categories")
plt.show()
```

## Univariate Insights

- Several numerical features show right-skewed distributions.
- The target variable `median_house_value` appears to have a capped upper limit.
- Outliers are present in features such as `total_rooms` and `population`.
- The categorical feature `ocean_proximity` is imbalanced across categories.
